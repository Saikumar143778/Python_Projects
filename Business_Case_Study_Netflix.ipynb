```python
def remove_space(x):
    if x!='':
        return str(x).strip()
```

```python
constraint_cast=df['cast'].apply(lambda x: str(x).split(',')).to_list()
```

```python
constraint_cast
```

```python
cast_df=pd.DataFrame(constraint_cast,index=df['title'])
```

```python
cast_df
```

```python
cast_df_stacked=cast_df.stack()
```

```python
cast_df_stacked=pd.DataFrame(cast_df_stacked.reset_index())
```

```python
cast_df_stacked.rename({0:'Actors'},axis=1,inplace=True)
```

```python
cast_df_stacked.drop(['level_1'],axis=1,inplace=True)
```

```python
cast_df_stacked
```

```python
cast_df_stacked['Actors']=cast_df_stacked['Actors'].apply(lambda x : remove_space(x))
```

```python
cast_df_stacked['Actors'].replace({'nan':'Unknown Actor',None:'Unknown Actor'}, inplace=True)
```

```python
cast_df_stacked
```

```python
constraint_director=df['director'].apply(lambda x: str(x).split(',')).to_list()
```

```python
director_df=pd.DataFrame(constraint_director,index=df['title'])
```

```python
director_df
```

```python
director_df_stacked=director_df.stack()
```

```python
director_df_stacked=pd.DataFrame(director_df_stacked.reset_index())
```

```python
director_df_stacked.drop('level_1',axis=1,inplace=True)
```

```python
director_df_stacked.rename({0:'Directors'},axis=1,inplace=True)
```

```python
director_df_stacked
```

```python
director_df_stacked['Directors']=director_df_stacked['Directors'].apply(lambda x :remove_space(x))
```

```python
director_df_stacked['Directors'].replace({'nan':'Unknown Director',None:'Unknown Director'}, inplace=True)
```

```python
director_df_stacked
```

```python
constraint_genre=df['listed_in'].apply(lambda x: str(x).split(',')).to_list()
genre_df=pd.DataFrame(constraint_genre,index=df['title'])
genre_df_stacked=genre_df.stack()
genre_df_stacked=pd.DataFrame(genre_df_stacked.reset_index())
genre_df_stacked.rename({0:'Genre'},axis=1,inplace=True)
genre_df_stacked.drop('level_1',axis=1,inplace=True)
genre_df_stacked
```

```python
genre_df_stacked['Genre']=genre_df_stacked['Genre'].apply(lambda x : remove_space(x))
```

```python
constraint_country=df['country'].apply(lambda x: str(x).split(',')).to_list()
country_df=pd.DataFrame(constraint_country,index=df['title'])
country_df_stacked=country_df.stack()
country_df_stacked=pd.DataFrame(country_df_stacked.reset_index())
country_df_stacked.rename({0:'Country'},axis=1,inplace=True)
country_df_stacked.drop('level_1',axis=1,inplace=True)
country_df_stacked
```

```python
country_df_stacked['Country']=country_df_stacked['Country'].apply(lambda x : remove_space(x))
```

```python
country_df_stacked['Country'].replace({'nan':'Unknown Country',None:'Unknown Country'}, inplace=True)
```

```python
country_df_stacked
```

```python
df_final['Country']=df_final['Country'].apply(formatting)
```
